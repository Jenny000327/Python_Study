# 오일러의 곱셈 공식 원주율 구하기

```python
import math

pi = 0

for i in range(1,100000): #무한대 수는 큰값 100000으로 대체체
  pi += 1/pow(i,2)
print(math.sqrt(6*pi))
```

# 1000이하의 소수(prime number) 구하기

```python
import math

p = list(range(1,1001)) #1부터 1000까지 가지고 있는 prime 리스트 생성
for i in range(1,int(math.sqrt(1000))+1): #1부터 루트 1000까지 (제곱근까지만 돌려도 소수 판별 가능)
  for j in p:
    if not j % i and j // i >= 2 : 
      p.remove(j)
print(len(p))
```

```python
p = list(range(2,1001)) #1은 소수가 아니기 때문에 2부터 1000까지 
for i in range(2, int(math.sqrt(1000))+1):
  for j in p[:]: 
    if not j % i and j != i: #j가 i 로 나누어 떨어지고, i 가 j와 같지 않을 때 
      p.remove(j)
print(len(p)) #리스트 p 의 길이 반환
```
